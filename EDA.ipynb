#Plot different names trends on same line graph 

```python
new_df = df.pivot_table(values='count', index='year', columns='name')
try_df = new_df.ix[:,:10]
try_df.plot()
```

```python
try_df = new_df.ix[:,:10]
last_df = try_df[-15:]
last_df.plot()
```

```python
sample_df = new_df.sample(n=15, axis =1)
sample_df
```

```python
# Get current size
fig_size = plt.rcParams["figure.figsize"]
 
# Prints: [8.0, 6.0]
print "Current size:", fig_size
 
# Set figure width to 12 and height to 9
fig_size[0] = 12
fig_size[1] = 9
plt.rcParams["figure.figsize"] = fig_size
```
